```python
import pandas as pd
import torch
from torch import nn
from pandas import DataFrame
import os

data = pd.read_csv('/Users/xDAyN/Desktop/cs project/School Project/Dimuon_DoubleMu.csv')

class_names = {
    0: "J/ψ",
    1: "ψ(2S)",
    2: "ϒ(1S)",
    3: "ϒ(2S)",
    4: "ϒ(3S)",
    5: "Z boson",
    6: "Background"
}


def classify_mass(mass):
    if 3.0 <= mass <= 3.2:
        return 0  # J/ψ
    elif 3.6 <= mass <= 3.8:
        return 1  # ψ(2S)
    elif 9.3 <= mass <= 9.6:
        return 2  # ϒ(1S)
    elif 9.9 <= mass <= 10.1:
        return 3  # ϒ(2S)
    elif 10.3 <= mass <= 10.4:
        return 4  # ϒ(3S)
    elif 88 <= mass <= 94:
        return 5  # Z boson
    else:
        return 6  # Background or unknown

data['label'] = data['M'].apply(classify_mass)

print("Label distribution:")
print(data['label'].value_counts())

# Map numeric labels to class names
data['class_name'] = data['label'].map(class_names)
print("Label distribution:")
print(data['label'].value_counts())


train_length = int(0.8*len(data))
print(train_length)

train_data = data[:train_length]
print(train_data)

test_data = data[train_length:]
print(test_data)

train_data.to_csv('/Users/xDAyN/Desktop/cs project/School Project/data/train data')
test_data.to_csv('/Users/xDAyN/Desktop/cs project/School Project/data/test data')

# Define columns
cols = ['E1','px1','py1','pz1','pt1','eta1','phi1','E2','px2','py2','pz2','pt2','eta2','phi2']  # 15 features, removed duplicate pz1/pz2
X_train = torch.tensor(train_data[cols].values, dtype=torch.float32)
X_test = torch.tensor(test_data[cols].values, dtype=torch.float32)
y_train = torch.tensor(train_data['label'].values, dtype=torch.long).unsqueeze(1)
y_test = torch.tensor(test_data['label'].values, dtype=torch.long).unsqueeze(1)


# Save to .pt files
os.makedirs("data", exist_ok=True)
torch.save(X_train, 'data/X_train.pt')
torch.save(y_train, 'data/y_train.pt')
torch.save(X_test, 'data/X_test.pt')
torch.save(y_test, 'data/y_test.pt')
```
